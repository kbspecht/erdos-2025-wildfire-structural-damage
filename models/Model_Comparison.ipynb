Let's compare the different models we have created. We have 7 models to compare: A naive model, a Bayes classifier, logistic regression, knn, a decision tree, random forest, and gradient boosting. 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set_style("whitegrid")
```

```python
Models = np.array(['Baseline', 'Naive Bayes', 'Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'Gradient Boosting'])
Accuracies = np.array([0.5976635514018691, 0.831385837812573, 0.8408506660434681, 0.8615330684739425, 0.8836955675001947, 0.8628184155176443, 0.8749707875671886])
Recalls = np.array([1.000000000, 0.7345338367702585, 0.68051118, 0.8755821689953427, 0.83, 0.7255300609933197, 0.76822538])
Precisions = np.array([0.59766355, 0.8270111183780249, 0.89942418, 0.8755821689953427, 0.88, 0.916024935826916, 0.90675351])
F1s = np.array([0.74817198, 0.780000, 0.77480159, 0.8162505814855017, 0.85,0.8097244732576986, 0.83176101, ])
```

```python
plt.figure(figsize=(12,7))
plt.bar(Models, Accuracies)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy Score", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12,7))

plt.plot(Models, Accuracies, 'o', label = 'Accuracy Scores' )
plt.plot(Models, Precisions, 'o', label = 'Precision Scores')
plt.plot(Models, Recalls, 'o', label = 'Recall Scores')
plt.plot(Models, F1s, 'o', label = 'F1 Scores')
plt.legend(loc = 'upper center')

plt.plot(Models, Accuracies*0 + np.max(Accuracies),'--', color = 'blue')
plt.plot(Models, Accuracies*0 + np.max(Precisions),'--', color = 'orange')
plt.plot(Models, Accuracies*0 + np.max(Recalls),'--', color = 'green')
plt.plot(Models, Accuracies*0 + np.max(F1s),'--', color = 'red')
plt.plot([],[],'--', label = 'Maximum Scores', color = 'black')
plt.legend(loc = 'upper center')



plt.xticks(fontsize=10)
plt.yticks([.6 + .05*i for i in range(9)],fontsize=10)


plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy Score", fontsize=12)

plt.show()
```
