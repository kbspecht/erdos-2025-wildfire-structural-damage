```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.preprocessing import LabelEncoder
import copy

features = []
min_loss = np.nan
min_loss_index = 0
les = []
cols = fire_train.select_dtypes('object').columns
for col in cols:
    le = LabelEncoder()
    fire_train[col] = le.fit_transform(fire_train[col])
    les.append(le)

for f in range(len(fire_train.columns)-1):
    fire_train_edited = fire_train.drop(columns=features)
    fire_train_edited = fire_train_edited.drop(columns='Destroyed')
    log_reg_accs = np.zeros((5, len(fire_train_edited.columns)))
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=216)
    i=0
    for train_index, test_index in kfold.split(fire_train, fire_train.Destroyed):
        fire_tt = fire_train.iloc[train_index]
        fire_ho = fire_train.iloc[test_index]

        j=0
        for feature in fire_train_edited.columns:
            if not features:
                lr = LogisticRegression(penalty=None)
                lr.fit(fire_tt[feature].values.reshape(-1,1),
                            fire_tt.Destroyed.values)
                pred = lr.predict(fire_ho[feature].values.reshape(-1,1))
            else:
                lr = LogisticRegression(penalty=None, max_iter=1000)
                new_features = copy.deepcopy(features)
                new_features.append(feature)
                lr.fit(fire_tt[new_features], fire_tt.Destroyed.values)
                pred = lr.predict(fire_ho[new_features])
            log_reg_accs[i,j] = log_loss(fire_ho.Destroyed.values,
                                                pred)        
            j = j + 1
        i = i + 1

    i=0
    min_loss_index = -1
    for loss in np.mean(log_reg_accs, axis = 0):
        if np.isnan(min_loss):
            min_loss = loss
            min_loss_index = i
        else:
            if loss < min_loss:
                min_loss = loss
                min_loss_index = i
        i += 1
    if min_loss_index == -1:
        break
    features.append(fire_train_edited.columns[min_loss_index])
    print(features)
```
