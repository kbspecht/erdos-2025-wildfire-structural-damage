Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

We start by reading data in from the cal_fire_data_cleaned CSV file:

```python
cal_fire_data = pd.read_csv("../data/cal_fire_data_cleaned.csv")
cal_fire_data.info()
```

We create a train/test split (80%/20%) for the data:

```python
from sklearn.model_selection import train_test_split
fire_train, fire_test = train_test_split(cal_fire_data, test_size=0.2, shuffle=True, random_state=216, stratify=cal_fire_data['Destroyed'])
fire_train['Destroyed'].value_counts(normalize=True)
```

We create a frequency histogram for values of Structure Type for each target class:

```python
print(cal_fire_data['* Structure Type'].value_counts())

plt.figure(figsize=[8,6], facecolor = ('#fffbf0'))   
ax = plt.gca()
ax.set_facecolor('#fffbf0')   
fire_train.loc[fire_train.Destroyed==0]['* Structure Type'].value_counts().sort_index().plot(kind='bar',
                                                                                color='blue',
                                                                                alpha=.4,
                                                                                hatch = '\\',
                                                                                edgecolor='black',
                                                                                label="$No Damage$")
fire_train.loc[fire_train.Destroyed==1]['* Structure Type'].value_counts().sort_index().plot(kind='bar',
                                                                                color='red',
                                                                                alpha=.4,
                                                                                hatch = '\\',
                                                                                edgecolor='black',
                                                                                label="$Damage$")
plt.title(f'Histogram of {'* Structure Type'}')
plt.xlabel('* Structure Type')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
